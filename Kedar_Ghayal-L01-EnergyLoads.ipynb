```python
# Import libraries
import pandas as pd
import numpy as np
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Cleaned up Dataset location
fileName = "https://library.startlearninglabs.uw.edu/DATASCI410/Datasets/EnergyEfficiencyData.csv"
```

```python
# Function to read data into a data frame.
def read_bldg_data(fileName):
    'Function to load the building data set from a .csv file' 
    ## Read the .csv file with the pandas read_csv method
    bldg_data = pd.read_csv(fileName)
    return bldg_data

bldg_data = read_bldg_data(fileName)
```

```python
# Create a scatter matrix to identify the relationships among the variables.
scatter_matrix(bldg_data,figsize  = [20, 20],diagonal = 'kde',alpha=0.5)
```

```python
# Examine the first and last few records.
bldg_data.head()
```

```python
# Examine the first and last few records.
bldg_data.tail()
```

```python
# Identify the # of records and the data types of each column
print(bldg_data.shape)
print(bldg_data.dtypes)
```

```python
#Identify the min , max, std,mean,and quartiles of each column.
bldg_data.describe()
```

```python
"""Below code is to examine the distinct values of each of the variables. 
It is found that Heating load & cooling load are indeed continuous & hence omitted below"""

for column in bldg_data.columns:
    if column in ['Heating Load','Cooling Load']:
        continue
    print(column, np.sort(bldg_data[column].unique()))
```

```python
""" In this section we plot the box plot of heating load and cooling load grouped by each of the 8 variables above
"""
fig,ax = plt.subplots(8,2,figsize=(8,32), dpi= 100)
k = np.array(range(0,8)) # counter for loop
column_list = []
for column in bldg_data.columns: # List of columns except the heating & cooling loads
    if column in ['Heating Load','Cooling Load']:
        continue
    column_list.append(column)
tuple = zip(k, column_list)
for k ,column in tuple:
    bldg_data.loc[:,['Heating Load', column]].boxplot(by = column, ax = ax[k,0])
    ax[k,0].set_xticklabels(ax[k,0].get_xticklabels(), rotation=270)
    bldg_data.loc[:,['Cooling Load', column]].boxplot(by = column, ax = ax[k,1])
    ax[k,1].set_xticklabels(ax[k,1].get_xticklabels(), rotation=270)
    plt.tight_layout()
    plt.suptitle('')
    plt.subplots_adjust(top = 2,bottom = 1)
    ax[k,0].set_ylim(np.min(bldg_data['Heating Load']), np.max(bldg_data['Heating Load'])) # Set the limits of the y axis
    ax[k,1].set_ylim(np.min(bldg_data['Cooling Load']), np.max(bldg_data['Cooling Load'])) # Set the limits of the y axis
```
